```python
phonebook={}
for i in range(10):
    name=input(str(i+1)+". 이름을 입력하세요: ")
    num=int(input(str(i+1)+". 전화번호를 입력하세요: "))
    print("")
    phonebook.update({name:num})
print("전화번호 사전 생성 완료\n")

while True:
    print("메뉴를 선택하세요: ")
    print("1. 이름으로 검색")
    print("2. 전화번호로 검색")
    print("3. 종료")
    menu=int(input())
    if(menu==1):
        found = False
        name=str(input("검색하고 싶은 이름을 입력하세요: "))
        for n, tel in phonebook.items(): 
            if name == n:
                print("이름: ",name)
                print("전화번호: ",str(tel))
                print("")
                found=not found
        if(not found):
            print("이름 존재하지 않음\n")
    elif(menu==2):
        found = False
        num=int(input("검색하고 싶은 전화번호를 입력하세요: "))
        for name, tel in phonebook.items(): 
            if tel == num:
                print("이름: ",name)
                print("전화번호: ",str(tel))
                print("")
                found=not found
        if(not found):
            print("전화번호 존재하지 않음\n")
    elif(menu==3):
        print("프로그램 종료")
        break
```
