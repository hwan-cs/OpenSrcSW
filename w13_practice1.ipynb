```python
import matplotlib.pyplot as plt
import csv
f=open('/Users/19juhpark/Downloads/bday_seoultemp.csv')
data=csv.reader(f)
next(data)
maxTemp=[]
minTemp=[]
for row in data:
    if row[-1] != '':
        if row[0].split('/')[1]=='10' and row[0].split('/')[2]=='05':
            maxTemp.append(float(row[-1]))
            minTemp.append(float(row[-2]))
plt.plot(maxTemp, color='skyblue')
plt.plot(minTemp, color='pink')
plt.show()
```
